```python
from flask import Flask , render_template, flash, request, redirect, url_for,session
import os
import cv2
from PIL import Image
import requests
import xmltodict
import json
from werkzeug.utils import secure_filename
from flask_ngrok import run_with_ngrok

UPLOAD_FOLDER = 'images'
app=Flask(__name__,static_url_path='/static')
run_with_ngrok(app)

app.secret_key = "1234"
app.config['UPLOAD_FOLDER'] = UPLOAD_FOLDER
ALLOWED_EXTENSIONS = set(['png', 'jpg', 'jpeg'])

 

def allowed_file(filename):
    return '.' in filename and \
           filename.rsplit('.', 1)[1].lower() in ALLOWED_EXTENSIONS

@app.route('/', methods=['GET', 'POST'])
def home_page():
    if request.method == 'POST':
        if 'car_image' not in request.files:
            flash('No file part')
            return redirect(request.url)
        file = request.files['car_image']

        if file.filename == '':
            flash('No selected file')
            return redirect(request.url)
        if file and allowed_file(file.filename):
            print("Image saved!!")
            print(file.filename)
            filename = secure_filename(file.filename)
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
            
    return render_template('home_screen.html')           

@app.route('/vehicle_data', methods=['GET', 'POST'])
def upload_file():
    if request.method == 'POST':
        
        if 'car_image' not in request.files:
            flash('No file part')
            return redirect(request.url)
        file = request.files['car_image']
        
        if file.filename == '':
            flash('No selected file')
            return redirect(request.url)
        if file and allowed_file(file.filename):
            print("Image saved!!")
            print(file.filename)
            filename = secure_filename(file.filename)
            file_extension=file_extension_finder(filename)
            filename="car."+file_extension
            file.save(os.path.join(app.config['UPLOAD_FOLDER'], filename))
            vehicle_number=no_plate_and_ocr_finder(filename)
            print(vehicle_number)
            if vehicle_number=="No vehicle Plate Found":
                print("No vehicle Plate Found")
                return  render_template('vehicle_data.html',error="No vehicle Plate Found")
            else:
                details=vehicle_info_finder(vehicle_number)
                if details=="No Info Found":
                    return render_template('vehicle_data.html',error="No Info Found")
                else:
                    return render_template(
                 "vehicle_data.html",
                 model_image = details[0],
                 model = details[1],
                 owner = details[2],
                 regyear = details[3],
                 engsize = details[4],
                 seats = details[5],
                 idfi = details[6],
                 engnum = details[7],
                 fuel_type = details[8],
                 vehicle_type = details[9],
                 regdate = details[10],
                 loc = details[11]
         )

def file_extension_finder(filename):
    return '.' in filename and \
           filename.rsplit('.', 1)[1].lower()

def no_plate_and_ocr_finder(filename):
     print(filename)
     file_location='/content/images/'+filename
     test_car=cv2.imread(file_location)
     gray = cv2.cvtColor(test_car, cv2.COLOR_BGR2GRAY)
     bfilter = cv2.bilateralFilter(gray, 11, 17, 17)
     edged = cv2.Canny(bfilter, 30, 200)
     keypoints = cv2.findContours(edged.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
     contours = imutils.grab_contours(keypoints)
     contours = sorted(contours, key=cv2.contourArea, reverse=True)[:10]

     location = None
     for contour in contours:
      approx = cv2.approxPolyDP(contour, 10, True)
      if len(approx) == 4:
       location = approx
       break
     
     mask = np.zeros(gray.shape, np.uint8)
     new_image = cv2.drawContours(mask, [location], 0,255, -1)
     new_image = cv2.bitwise_and(test_car, test_car, mask=mask)

     (x,y) = np.where(mask==255)
     (x1, y1) = (np.min(x), np.min(y))
     (x2, y2) = (np.max(x), np.max(y))
     cropped_image = gray[x1:x2+1, y1:y2+1]

     reader = easyocr.Reader(['en'])
     result = reader.readtext(cropped_image)
     print(result)
    
     plate_number = result[0][-2]
     plate_number = plate_number.replace(' ', '')
     plate_number = plate_number.replace('*','')
     return plate_number

def vehicle_info_finder(plate_number):
    try:
        r = requests.get("http://www.regcheck.org.uk/api/reg.asmx/CheckIndia?RegistrationNumber={0}&username=Sharma2666".format(str(plate_number)))
        data = xmltodict.parse(r.content)
        jdata = json.dumps(data)
        df = json.loads(jdata)
        df1 = json.loads(df['Vehicle']['vehicleJson'])
        print(df1)
        return [df1["ImageUrl"],
          df1["Description"],
          df1["Owner"],
          df1["RegistrationYear"],
          df1["EngineSize"]["CurrentTextValue"],
          df1["NumberOfSeats"]["CurrentTextValue"],
          df1["VechileIdentificationNumber"],
          df1["EngineNumber"],
          df1["FuelType"]["CurrentTextValue"],
          df1["VehicleType"],
          df1["RegistrationDate"],
          df1["Location"]]
    except:
        return "No Info Found"

 
if __name__ == '__main__':
    app.run()
```
